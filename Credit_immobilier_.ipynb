## 1.2 : Création d'une base de données 

```python
#Définition des variables constitutives du tableau 
caracteristiques = ['taux_de_ventes', 'croissance_ventes', 'ratio_benefices', 'ratio_pertes']
#On veut faire en sorte que les variables 'taux_de_ventes' et 'croissance_ventes' soient comprises entre 0 et 100
#Parallèlement, on veut que les variables 'ratio_benefice' et 'ratio_pertes' soient comprises entre 0 et 1
#On utilise la formule "np.random.rand" pour cela. ".rand" créé des valeurs entre 0.0 et 1.0. Donc: 
variable1 = np.random.rand(6, 2)*100
variable2 = np.random.rand (6,2)
#On concatène les deux variables créées pour qu'elles se retrouvent dans le même tableau, à 6 lignes et 4 colonnes
tableau = np.column_stack((variable1, variable2))
#On affiche le tableau
df = pd.DataFrame(tableau, columns=list(caracteristiques), index=['1', '3', '4', '6', '7', '9'])
df
```

# 3. Traitement des données

## 1.1 : Importation des bibliothèques nécessaires 

```python
import pandas as pd 
import numpy as np
#Sklearn est un outil d'apprentissage automatique
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## 1.2 Importer les jeux de données

```python
df = pd.read_csv("Data/credit_immo/credit_immo.csv")
df
```

## 1.3 Transformer les valeurs manquantes en moyenne

```python
#Premièrement, à l'aide de la fonction '.dtypes', on va exlcure les colonnes contenant des strings
col= df.select_dtypes(exclude='object').columns
col
```

```python
#SimpleImputer à pour objectif de chercher les valeurs manquantes, ou NaN, en lui appliquant une stratégie, ici une moyenne
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

#Ensuite, on va dire à SimpleImputer où faire cette opération, ici dans notre dataframe 'df'
imputer.fit(df[col])

#La dernière fonction sert à afficher le résultat du calcul
df[col] = imputer.transform(df[col])

#Enfin, on affiche le tableau
df
```

## 1.4 Encoder les valeurs catégoriques (LabelEncoder)

```python
df2 = df.apply(LabelEncoder().fit_transform)
df2
```

## 1.5 Fractionner le jeu de données pour l’entrainement et le test (Training and Test set)

```python
#On sépare le jeu de données en deux mesures, X et y:
X = df.iloc[: , 0:-1]
y = df["Solvable"]
#La formule qui servira de de phase d'entraînement et de test est:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```
